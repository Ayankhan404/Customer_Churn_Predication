#### Data preprocessing

```python
# Label encoding of targget column
df['Churn']=df['Churn'].replace({'Yes':1,'No':0})
```

```python
#Label encoding on categorical column
categorial_fearues=df.select_dtypes('object').columns
categorial_fearues
```

```python
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
encoder={}

#apply lable encoder and store the encoders
for col in categorial_fearues:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    encoder[col]=le

#save the encoder to pickle file 
with open("encoder.pkl","wb") as f:
    pickle.dump(encoder,f)   
```

```python
x=df.drop(columns=['Churn'])
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
## handling Imbalaced data
from imblearn.over_sampling import SMOTE

oversample=SMOTE(random_state=42)
x_train_smote,y_train_smote = oversample.fit_resample(x_train,y_train)
```

```python
print(y_train_smote.shape)
print(x_train_smote.shape)
```

```python
print(y_train_smote.value_counts())
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix ,classification_report
```

```python
tr=tree.DecisionTreeClassifier(random_state=42)
rf=RandomForestClassifier(random_state=42)
xb=XGBClassifier(random_state=42)
```

```python
y_pred_tr=tr.predict(x_test)
y_pred_tr
```

```python
report1=classification_report(y_test,y_pred_tr)
print(report1)
```

```python
tr.score(x_test,y_test)
```

```python
# save the trained model as a pickle file
model_data = {"model": xb, "features_names": x.columns.tolist()}


with open("customer_churn_model.pkl", "wb") as f:
  pickle.dump(model_data, f)
```

```python
# load teh saved model and the feature names

with open("customer_churn_model.pkl", "rb") as f:
  model_data = pickle.load(f)

loaded_model = model_data["model"]
feature_names = model_data["features_names"]
```

```python
input_data = {
    'gender': 'Female',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'No',
    'tenure': 1,
    'PhoneService': 'No',
    'MultipleLines': 'No phone service',
    'InternetService': 'DSL',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 29.85,
    'TotalCharges': 29.85
}


input_data_df = pd.DataFrame([input_data])

with open("encoder.pkl", "rb") as f:
  encoder = pickle.load(f)


# encode categorical featires using teh saved encoders
for column, encoder in encoder.items():
  input_data_df[column] = encoder.transform(input_data_df[column])

# make a prediction
prediction = loaded_model.predict(input_data_df)
pred_prob = loaded_model.predict_proba(input_data_df)

print(prediction)

# results
print(f"Prediction: {'Churn' if prediction[0] == 1 else 'No Churn'}")
print(f"Prediciton Probability: {pred_prob}")
```
